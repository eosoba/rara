```python
from sklearn.cross_validation import train_test_split
a_train, a_test = train_test_split(ter, test_size=0.3, random_state=42)
```

```python

linearfit=smf.ols(formula='MEDV ~ LAT + LON + CRIM + ZN + INDUS + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO',data=a_train).fit()
print linearfit.summary()
```

```python
y,xlintest = dmatrices('MEDV ~ LAT + LON + CRIM + ZN + INDUS + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO', data=a_train,return_type='dataframe')
```

```python
ypred=linearfit.predict(xlintest)
```

```python
print sum((a_test['MEDV']-ypred)**2)
mean_squared_error(y.as_matrix(),ypred)*152
```

```python
y,X= dmatrices('MEDV ~ LAT + LON + CRIM + ZN + INDUS + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO-1',a_train,return_type='dataframe')

clf = tree.DecisionTreeRegressor()
y = np.ravel(y)
clf = clf.fit(X, y)

ylog,xlogtest = dmatrices('MEDV ~ LAT + LON + CRIM + ZN + INDUS + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO-1', data=a_test,return_type='dataframe')
ylog=np.ravel(ylog)
ypred=clf.predict(xlogtest)
mean_squared_error(ylog,ypred)*152
```
