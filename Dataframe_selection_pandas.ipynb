# Various methods to filter a dataframe

```python
from IPython.display import HTML, display

def set_background(color):    
    script = (
        "var cell = this.closest('.jp-CodeCell');"
        "var editor = cell.querySelector('.jp-Editor');"
        "editor.style.background='{}';"
        "this.parentNode.removeChild(this)"
    ).format(color)
    
    display(HTML('<img src onerror="{}" style="display:none">'.format(script)))
set_background('honeydew')
```

```python
import pandas as pd
df = pd.read_csv("healthcare-dataset-stroke-data.csv") 
print(df.columns)
df.head()
```

## Select individual columns

```python
df['id'].head()
```

## Select multiple columns

```python
df[['hypertension','age']].head()
```

##  List based methods

```python
columns = df.columns.tolist()
df[columns[4:7]].head()
```

## filter based on a criterion
- subset based on equality

```python
df[df['ever_married'] == 'Yes'].head()
```

## Subset based on existence in a list

```python
df[df['work_type'].isin(['Private','Govt_job'])].head()
```

## Subset based on numerical comparison

```python
df[df['hypertension'] > 0.5].head()
```

## Subsetting using more complex logic

```python
df[(df['hypertension'] > 0.5) & (df['gender'] == 'Female')].head()  # hypertension > 0.5 AND gender = Female
```
