'Drink', 'drinking', 'drunk' are all from the same verb.  When we are doing frequency analysis, it is sometimes crucial to further to reduce words to the same parent. This is known as stemming.  There are different methods for stemming.



```python
stemmer = nltk.PorterStemmer()
plurals = ['caresses', 'flies', 'dies', 'mules', 'denied', 'died', 'agreed', 'owned', 'humbled', 'sized', 'sensational', 'traditional', 'reference', 'colonizer','plotted']
singles = []
for plural in plurals:
    singles.append(stemmer.stem(plural))
print singles
```

An important part of nlap is sentencing tagging.  

```python
mytext = nltk.word_tokenize('My name is Crazy Fish')
nltk.pos_tag(mytext)
```

Now that we can do a part of speech tag, the next step is do some chunking. 
Noun phrase chunking.
1. (We) saw (the yellow dog). (2 NP chunks, parenthesis)
2. (The market) for (system-management software) for 
(Digital's hardware) is fragmented enough that (a giant) such as 
(Computer Associates) should do well there. (5 NP chunks)

```python
nltk.ne_chunk(nltk.pos_tag(mytext))
nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize('San Francisco is always rainy')))
```

```python
# We can chunk using regex parser by defining a grammar.
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"), ("dog", "NN"), ("barked", 
"VBD"), ("at", "IN"), ("the", "DT"), ("cat", "NN")] # a simple sentence with POS tags
pattern = "NP: {<DT>?<JJ>*<NN>}" # define a tag pattern of an NP chunk
NPChunker = nltk.RegexpParser(pattern) # create a chunk parser 
result = NPChunker . parse(sentence) # parse the example sentence
print result 
```

```python
colors = 'rgbcmyk' # red, green, blue, cyan, magenta, yellow, black
def bar_chart(categories, words, counts):
    ind = pylab.arange(len(words))
    width = 1 / (len(categories) + 1)
    bar_groups = []
    for c in range(len(categories)):
        bars = pylab.bar(ind+c*width, counts[categories[c]], width,
        color=colors[c % len(colors)])
        bar_groups.append(bars)
    pylab.xticks(ind+width, words)
    pylab.legend([b[0] for b in bar_groups], categories, loc='upper left')
    pylab.ylabel('Frequency')
    pylab.title('Frequency of Six Modal Verbs by Genre')
    pylab.show()
genres = ['news', 'religion', 'hobbies', 'government', 'adventure']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfdist = nltk.ConditionalFreqDist(
(genre, word)
for genre in genres
for word in nltk.corpus.brown.words(categories=genre)
if word in modals)

counts = {}
for genre in genres:
    counts[genre] = [cfdist[genre][word] for word in modals]
bar_chart(genres, modals, counts)
```
